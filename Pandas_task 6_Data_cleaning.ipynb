### Missing values

```python
df['Age'].fillna(df['Age'].median(),inplace = True)
df['Gender'].fillna(df['Gender'].mode()[0],inplace= True)
df['Item Purchased'].fillna(df['Item Purchased'].mode()[0],inplace= True)
df['Category'].fillna(df['Category'].mode()[0],inplace = True)
df['Location'].fillna(df['Location'].mode()[0],inplace= True)
df['Color'].fillna(df['Color'].mode()[0],inplace= True)
df['Season'].fillna(df['Season'].mode()[0],inplace= True)
df['Review Rating'].fillna(df['Review Rating'].median(),inplace= True)
df['Subscription Status'].fillna(df['Subscription Status'].mode()[0],inplace= True)
df['Discount Applied'].fillna(df['Discount Applied'].mode()[0],inplace= True)
df['Promo Code Used'].fillna(df['Promo Code Used'].mode()[0],inplace= True)
df['Payment Method'].fillna(df['Payment Method'].mode()[0],inplace= True)
df['Frequency of Purchases'].fillna(df['Frequency of Purchases'].mode()[0],inplace= True)
```

```python
df.isna().sum()
```

### outlier

```python
sum((df['Age']<0)|(df['Age']>100))
```

```python
condition = (df['Age']<0)|(df['Age']>100)

df['Age'] = np.where(condition,30,df['Age'])
condition1 = (df["Review Rating"] < 0) | (df["Review Rating"] > 5)  

df["Review Rating"] = np.where(condition1,4,df["Review Rating"])
```

### Scalling

```python
df.head(4)
```

```python
df['Age'].mean(),df['Age'].median()
```

```python
def sales(x):
    n = (x - df['Age'].mean())/ df['Age'].std()
    return n

df['Age'] = df['Age'].apply(sales)
```

```python
df["Purchase Amount (USD)"].mean(),df["Purchase Amount (USD)"].median()
```

```python
def sales(x):
    n = (x - df['Purchase Amount (USD)'].mean())/ df['Purchase Amount (USD)'].std()
    return n

df['Purchase Amount (USD)'] = df['Purchase Amount (USD)'].apply(sales)
```

```python
df['Review Rating'].mean(),df['Review Rating'].median()
```

```python
def sales(x):
    n = (x - df['Review Rating'].mean())/ df['Review Rating'].std()
    return n

df['Review Rating'] = df['Review Rating'].apply(sales)
```

```python
def sales(x):
    n = (x - df['Previous Purchases'].mean())/ df['Previous Purchases'].std()
    return n

df['Previous Purchases'] = df['Previous Purchases'].apply(sales)
```
