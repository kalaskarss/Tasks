### Stadardization

```python
df1['Quantity'] = df1['Quantity'].astype('int32')
```

```python
def standardization(x):
    n = (x - df1['Quantity'].mean())/df1['Quantity'].std()
    return n
```

```python
df1['Quantity'] = df1['Quantity'].apply(standardization)
df1['Quantity']
```

### Data Manipulation - concatenate and merge the data.

```python
pd.concat([df1,df2],axis = 0,ignore_index = True)
```

```python
pd.concat([df1,df2],axis = 1,ignore_index = True)
```

```python
pd.merge(df1,df2,on = "InvoiceNo") 
```

```python
pd.merge(df1,df2,on = "InvoiceNo",how = 'right') 
```

```python
pd.merge(df1,df2,on = "InvoiceNo",how = 'left') 
```
