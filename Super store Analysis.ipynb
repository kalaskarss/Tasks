### count plot

### Pie chart

```python
(df["Region"].value_counts()/len(df["Region"])*100).plot(kind = "pie",
                                                               ylabel = 'percentage',
                                                                figsize = (6,7),
                                                                autopct = '%1.1f%%')
```

```python
(df["Sub-Category"].value_counts()/len(df["Sub-Category"])*100).plot(kind = "pie",
                                                               ylabel = 'percentage',
                                                                figsize = (6,7),
                                                                autopct = '%1.1f%%')
```
